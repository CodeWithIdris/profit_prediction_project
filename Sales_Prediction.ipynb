```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import root_mean_squared_error, r2_score
from sklearn.model_selection import cross_val_score
import numpy as np
import joblib
```

```python
data = pd.read_csv(r"C:\Users\FAMILY\Downloads\Sales Prediction\SUPERSTORE.csv, encoding='ISO-8859-1')
data
```

```python
data.isnull().sum()
```

```python
data.isnull().sum()
```

```python
#select the features variable and the target

features = data[['Sales', 'Discount', 'Quantity', 'Category', 'Sub-Category']]
target = data['Profit']

features_encoded = pd.get_dummies(features, columns = ['Category', 'Sub-Category'], drop_first = True)

```

```python
x_train, x_test, y_train, y_test = train_test_split(features_encoded, target, test_size = 0.2, random_state = 42)

x_train.head()
```

```python
model = RandomForestRegressor(random_state = 42)
```

```python
#training the model using the training data

model.fit(x_train, y_train)
```

```python
#predicting the profit on the test data

y_prediction = model.predict(x_test)
```

```python
# Evaluate the model: calculate RMSE and R-squared

rmsr = root_mean_squared_error(y_test, y_prediction)

#R-squared
r2 = r2_score(y_test, y_prediction) 
```

```python
print(f"Root Mean Squared Error (RMSE): {rmsr}")
print(f"R-squared: {r2}")
```

```python
#check for overfitting

y_train_prediction = model.predict(x_train)
```

```python
#calculating the R sqaured and RMSE on the training data

train_rmse = root_mean_squared_error(y_train, y_train_prediction)
train_r2 = r2_score(y_train, y_train_prediction)
```

```python
y_prediction.shape
```

```python

test_rmse = root_mean_squared_error(y_test, y_prediction)
test_r2 = r2_score(y_test, y_prediction)
```

```python
print(f"Training RMSE: {train_rmse}")
print(f"Training R-squared: {train_r2}")
```

```python
# Calculating the R-squared and RMSE on the test data
test_rmse = root_mean_squared_error(y_test, y_prediction)
test_r2 = r2_score(y_test, y_prediction)

# Display the results
print(f"Test RMSE: {test_rmse}")
print(f"Test R-squared: {test_r2}")
```

```python
# Extract feature importances
feature_importances = model.feature_importances_

# Create a DataFrame for better visualization
features_df = pd.DataFrame({'Feature': x_train.columns, 'Importance': feature_importances})

# Sort features by importance
features_df = features_df.sort_values(by='Importance', ascending=False)
```

```python
# Plot feature importance
plt.figure(figsize=(10, 6))
plt.barh(features_df['Feature'], features_df['Importance'])
plt.xlabel('Importance')
plt.title('Feature Importance')
plt.gca().invert_yaxis()
```

```python
# Perform cross-validation on the model
cv_scores = cross_val_score(model, x_train, y_train, cv=5, scoring='neg_mean_squared_error')

# Calculate the mean and standard deviation of RMSE from cross-validation
cv_rmse = (-cv_scores) ** 0.5  # Taking the square root of negative MSE to get RMSE

print(f"Cross-Validation RMSE: Mean = {cv_rmse.mean()}, Std = {cv_rmse.std()}")
```

```python
# Residual analysis
residuals = y_test - y_prediction

plt.figure(figsize=(10, 6))
plt.scatter(y_prediction, residuals)
plt.axhline(0, color='red', linestyle='--')
plt.xlabel('Predicted Profit')
plt.ylabel('Residuals')
plt.title('Residual Plot')
plt.show()
```

```python
# Saving the model to a file
joblib.dump(model, 'random_forest_profit_model.pkl')
```
